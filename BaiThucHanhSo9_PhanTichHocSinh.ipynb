### Câu 1: Tỷ lệ Điểm Trung bình Cuối kỳ (G3) giữa các trường

**Mục tiêu:** Biểu đồ này giúp nhà quản lý giáo dục so sánh nhanh chóng mức độ thành công trung bình của hai trường (GP và MS) dựa trên điểm số cuối kỳ (G3) của học sinh. Biểu đồ sẽ thể hiện tỷ lệ phần trăm điểm trung bình của mỗi trường trong tổng điểm trung bình của cả hai.

**Các bước thực hiện:**
1.  Đọc dữ liệu từ file `student-mat.csv`.
2.  Nhóm dữ liệu theo từng trường (`school`).
3.  Tính điểm trung bình cuối kỳ (G3) cho mỗi trường.
4.  Vẽ biểu đồ tròn (pie chart) để trực quan hóa tỷ lệ phần trăm của điểm trung bình mỗi trường.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Thiết lập style cho biểu đồ
sns.set_style("whitegrid")

# Đọc dữ liệu
try:
    data = pd.read_csv('student-mat.csv')
except FileNotFoundError:
    print("Không tìm thấy file student-mat.csv")
    data = None

if data is not None:
    # 1. Tính điểm trung bình G3 cho mỗi trường
    school_avg_g3 = data.groupby('school')['G3'].mean().reset_index()

    # 2. Chuẩn bị dữ liệu cho biểu đồ tròn
    labels = school_avg_g3['school']
    sizes = school_avg_g3['G3']

    # 3. Vẽ biểu đồ tròn
    plt.figure(figsize=(8, 8))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
    plt.title('Tỷ lệ Điểm Trung bình Cuối kỳ (G3) giữa các Trường')
    plt.ylabel('') # Ẩn nhãn trục y không cần thiết
    plt.show()
```

### Diễn giải kết quả

Biểu đồ tròn ở trên thể hiện tỷ lệ phân chia điểm trung bình cuối kỳ (G3) giữa hai trường: Gabriel Pereira (GP) và Mousinho da Silveira (MS).

- **Phân tích:** Kết quả cho thấy điểm trung bình của hai trường khá tương đồng, với trường GP chiếm tỷ lệ nhỉnh hơn một chút so với trường MS. Cụ thể, trường GP chiếm khoảng 51.2% và trường MS chiếm khoảng 48.8% trong tổng điểm trung bình.
- **Ý nghĩa cho nhà quản lý:** Mặc dù có sự chênh lệch nhỏ, nhưng nhìn chung, kết quả học tập cuối kỳ trung bình của học sinh ở cả hai trường là khá cân bằng. Nhà quản lý có thể kết luận rằng không có sự khác biệt quá lớn về chất lượng đầu ra trung bình giữa hai trường này dựa trên điểm G3. Tuy nhiên, để có cái nhìn sâu hơn, cần phân tích thêm về sự phân phối điểm trong từng trường như ở câu hỏi tiếp theo.

### Câu 2: Biểu đồ phân phối điểm G1, G2, và G3 giữa các trường

**Mục tiêu:** Phân tích này giúp nhà quản lý hiểu rõ hơn về sự phân bổ điểm số của học sinh (G1, G2, và G3) trong từng trường. Thay vì chỉ nhìn vào điểm trung bình, biểu đồ phân phối cho thấy sự đa dạng trong kết quả học tập, ví dụ: học sinh của trường nào có điểm số đồng đều hơn, trường nào có nhiều học sinh đạt điểm cao hoặc thấp hơn.

**Các bước thực hiện:**
1.  Sử dụng dữ liệu đã đọc.
2.  Tạo ba biểu đồ, mỗi biểu đồ tương ứng với một kỳ thi (G1, G2, G3).
3.  Trong mỗi biểu đồ, vẽ biểu đồ mật độ (KDE plot) để thể hiện sự phân phối điểm số của từng trường.

```python
if data is not None:
    # Tạo một figure chứa 3 biểu đồ con (subplots)
    fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)
    fig.suptitle('Phân phối Điểm số các Kỳ thi giữa các Trường', fontsize=16)

    # Danh sách các kỳ thi
    grades = ['G1', 'G2', 'G3']

    # Vẽ biểu đồ cho mỗi kỳ thi
    for i, grade in enumerate(grades):
        sns.kdeplot(data=data, x=grade, hue='school', ax=axes[i], fill=True, common_norm=False)
        axes[i].set_title(f'Phân phối điểm {grade}')
        axes[i].set_xlabel('Điểm số')
        axes[i].set_ylabel('Mật độ')

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
```

### Diễn giải kết quả

Biểu đồ trên hiển thị sự phân phối điểm của học sinh qua ba kỳ thi (G1, G2, G3) tại hai trường GP và MS.

- **Phân tích:**
  - **Trường GP (Gabriel Pereira):** Phân phối điểm của trường GP có xu hướng tập trung ở mức cao hơn, đặc biệt là ở kỳ thi G3. Đỉnh của biểu đồ mật độ của trường GP thường nằm ở bên phải so với trường MS, cho thấy có một tỷ lệ lớn học sinh ở trường GP đạt điểm khá và giỏi.
  - **Trường MS (Mousinho da Silveira):** Phân phối điểm của trường MS có đỉnh thấp hơn và trải rộng hơn, cho thấy kết quả học tập của học sinh ở trường này đa dạng hơn, có cả học sinh điểm cao và một số lượng đáng kể học sinh ở mức điểm thấp hơn. Đặc biệt, có một đỉnh nhỏ ở khoảng điểm 0 ở kỳ thi G3, cho thấy có một nhóm học sinh bị điểm rất thấp.

- **Ý nghĩa cho nhà quản lý:**
  - Mặc dù điểm *trung bình* giữa hai trường không chênh lệch nhiều (như ở câu 1), biểu đồ phân phối này cho thấy bức tranh chi tiết hơn: trường GP có chất lượng học sinh đồng đều và tốt hơn ở nhóm đầu. 
  - Nhà quản lý của trường MS có thể cần quan tâm đến nhóm học sinh có kết quả học tập thấp, đặc biệt là nhóm bị điểm 0 ở kỳ G3, để tìm hiểu nguyên nhân và có phương án hỗ trợ kịp thời.

### Câu 3: Điểm Trung bình G3 theo Thời gian Học Tập Hàng Tuần

**Mục tiêu:** Đánh giá ảnh hưởng của thời gian học tập hàng tuần (`studytime`) đến điểm số cuối kỳ (G3). Thông qua biểu đồ, nhà quản lý có thể hiểu được liệu việc dành nhiều thời gian hơn cho việc học có thực sự giúp cải thiện kết quả học tập của học sinh hay không.

**Các bước thực hiện:**
1.  Nhóm dữ liệu theo các mức thời gian học tập (`studytime`).
2.  Tính điểm trung bình G3 cho mỗi nhóm.
3.  Sử dụng biểu đồ cột để so sánh điểm trung bình G3 giữa các nhóm.

```python
if data is not None:
    # Nhóm và tính điểm trung bình G3 theo thời gian học
    avg_g3_by_studytime = data.groupby('studytime')['G3'].mean().reset_index()

    # Ánh xạ các giá trị studytime sang nhãn có ý nghĩa hơn
    studytime_labels = {1: '< 2 giờ', 2: '2-5 giờ', 3: '5-10 giờ', 4: '> 10 giờ'}
    avg_g3_by_studytime['studytime_label'] = avg_g3_by_studytime['studytime'].map(studytime_labels)

    # Vẽ biểu đồ cột
    plt.figure(figsize=(10, 6))
    sns.barplot(data=avg_g3_by_studytime, x='studytime_label', y='G3', palette='viridis', order=['< 2 giờ', '2-5 giờ', '5-10 giờ', '> 10 giờ'])
    
    plt.title('Điểm Trung bình Cuối kỳ (G3) theo Thời gian Học Tập Hàng Tuần')
    plt.xlabel('Thời gian học hàng tuần')
    plt.ylabel('Điểm Trung bình G3')
    plt.ylim(0, 14) # Đặt giới hạn trục y để dễ so sánh
    plt.show()
```

### Diễn giải kết quả

Biểu đồ cột ở trên cho thấy mối quan hệ giữa thời gian học tập hàng tuần và điểm trung bình cuối kỳ (G3) của học sinh.

- **Phân tích:**
  - Nhìn chung, có một xu hướng tích cực: học sinh dành nhiều thời gian học tập hơn có xu hướng đạt điểm trung bình cao hơn. 
  - Điểm trung bình tăng đáng kể khi thời gian học tăng từ '< 2 giờ' lên '2-5 giờ' và '5-10 giờ'.
  - Tuy nhiên, một điểm đáng chú ý là nhóm học sinh học nhiều nhất ('> 10 giờ') lại có điểm trung bình thấp hơn một chút so với nhóm học '5-10 giờ'. Điều này có thể gợi ý rằng việc học quá nhiều không phải lúc nào cũng mang lại hiệu quả cao nhất, hoặc có thể nhóm học sinh này là những em đang gặp khó khăn trong học tập và phải dành nhiều thời gian hơn để theo kịp.

- **Ý nghĩa cho nhà quản lý:**
  - Khuyến khích học sinh dành một khoảng thời gian học tập hợp lý (ví dụ từ 2 đến 10 giờ mỗi tuần) là một chiến lược tốt để cải thiện kết quả.
  - Nhà quản lý cần lưu ý đến nhóm học sinh học trên 10 giờ mỗi tuần nhưng kết quả không tương xứng. Đây có thể là dấu hiệu của việc học không hiệu quả. Cần có các chương trình tư vấn, hướng dẫn phương pháp học tập phù hợp để giúp các em tối ưu hóa thời gian và công sức bỏ ra.

### Câu 4: Phân phối số lần nghỉ học (`absences`) giữa các trường

**Mục tiêu:** Giúp nhà quản lý hiểu rõ tình hình chuyên cần của học sinh giữa hai trường. Biểu đồ sẽ cho thấy sự phân bổ số buổi nghỉ học, giúp xác định xem trường nào có tỷ lệ học sinh nghỉ học nhiều hơn và liệu có những trường hợp nghỉ học cá biệt (outliers) hay không.

2.  Vẽ biểu đồ violin để so sánh sự phân phối của cột `absences` giữa hai trường (`school`). Biểu đồ violin kết hợp ưu điểm của box plot (hiển thị tứ phân vị) và KDE plot (hiển thị mật độ phân phối).

  - Mặc dù trung vị (đường trắng ở giữa) và phần lớn học sinh của hai trường có số buổi nghỉ tương tự nhau, nhưng sự tồn tại của các trường hợp cá biệt ở trường GP là một điểm khác biệt.

Cần tìm hiểu nguyên nhân đằng sau (lý do sức khỏe, vấn đề gia đình, hay các vấn đề khác) để có biện pháp can thiệp và hỗ trợ phù hợp, vì việc nghỉ học quá nhiều có thể ảnh hưởng tiêu cực đến kết quả học tập.

### Câu 5: Ảnh hưởng của Thời gian Tự do (`freetime`) đến Điểm số

**Mục tiêu:** Kiểm tra xem thời gian rảnh rỗi của học sinh có ảnh hưởng đến kết quả học tập cuối kỳ (G3) hay không. Thông tin này có thể giúp nhà quản lý xây dựng các chương trình hoạt động ngoại khóa, tư vấn về việc cân bằng giữa học tập và giải trí.

**Các bước thực hiện:**
1.  Nhóm dữ liệu theo các mức độ thời gian rảnh (`freetime`).
2.  Tính điểm trung bình G3 cho mỗi nhóm.
3.  Vẽ biểu đồ cột để so sánh điểm trung bình.

```python
if data is not None:
    # Nhóm và tính điểm trung bình G3 theo thời gian rảnh
    avg_g3_by_freetime = data.groupby('freetime')['G3'].mean().reset_index()
    
    # Ánh xạ các giá trị freetime sang nhãn có ý nghĩa
    freetime_labels = {1: 'Rất ít', 2: 'Ít', 3: 'Trung bình', 4: 'Nhiều', 5: 'Rất nhiều'}
    avg_g3_by_freetime['freetime_label'] = avg_g3_by_freetime['freetime'].map(freetime_labels)
    
    # Vẽ biểu đồ cột
    plt.figure(figsize=(10, 6))
    sns.barplot(data=avg_g3_by_freetime, x='freetime_label', y='G3', palette='plasma', order=['Rất ít', 'Ít', 'Trung bình', 'Nhiều', 'Rất nhiều'])
    
    plt.title('Điểm Trung bình G3 theo Mức độ Thời gian Tự do')
    plt.xlabel('Thời gian tự do sau giờ học')
    plt.ylabel('Điểm Trung bình G3')
    plt.ylim(0, 13)
    plt.show()
```

Biểu đồ thể hiện mối liên hệ giữa lượng thời gian tự do của học sinh và điểm trung bình cuối kỳ (G3).

- **Phân tích:** Có một xu hướng khá bất ngờ và thú vị: những học sinh có *ít* thời gian rảnh hơn (mức 1 và 2) lại có xu hướng đạt điểm trung bình G3 cao hơn so với những học sinh có *nhiều* thời gian rảnh (mức 4 và 5). Nhóm học sinh có thời gian rảnh ở mức trung bình (mức 3) có điểm số khá tốt, nhưng vẫn thấp hơn nhóm có ít thời gian rảnh.

- **Ý nghĩa cho nhà quản lý:**
  - Điều này không có nghĩa là thời gian rảnh gây hại. Thay vào đó, nó có thể ngụ ý rằng những học sinh có kết quả học tập tốt hơn thường là những em có kỷ luật, dành nhiều thời gian hơn cho việc học hoặc các hoạt động có cấu trúc khác (như học thêm, tham gia câu lạc bộ học thuật), dẫn đến việc có ít thời gian rảnh rỗi hơn.
  - Nhà trường có thể xem xét việc tổ chức các hoạt động ngoại khóa, câu lạc bộ có định hướng và mục tiêu rõ ràng. Điều này vừa giúp học sinh sử dụng thời gian rảnh một cách hiệu quả, vừa phát triển kỹ năng mà không làm ảnh hưởng tiêu cực đến kết quả học tập như khi có quá nhiều thời gian rảnh không định hướng.

### Câu 6: Biểu đồ Tọa độ Song song: Mối quan hệ giữa Điểm số, Thời gian học, Sức khỏe và Mức tiêu thụ rượu

**Mục tiêu:** Giúp nhà quản lý nhìn thấy bức tranh tổng thể về mối liên hệ giữa nhiều yếu tố khác nhau của học sinh. Biểu đồ này cho phép so sánh đồng thời các chỉ số như điểm cuối kỳ (G3), thời gian học, tình trạng sức khỏe, và mức tiêu thụ rượu trong ngày đi học (Dalc) và cuối tuần (Walc) để nhận diện các nhóm học sinh hoặc các xu hướng đặc biệt.

**Các bước thực hiện:**
1.  Chọn ra các cột dữ liệu cần thiết: `G3`, `studytime`, `health`, `Dalc`, `Walc` và `school` (để phân biệt).
2.  Sử dụng chức năng `parallel_coordinates` của Pandas để vẽ biểu đồ.
3.  Mỗi đường trên biểu đồ đại diện cho một học sinh, đi qua các trục tương ứng với các chỉ số.

```python
if data is not None:
    # Chọn các cột cần thiết cho phân tích
    parallel_data = data[['school', 'G3', 'studytime', 'health', 'Dalc', 'Walc']]
    
    # Vẽ biểu đồ tọa độ song song
    plt.figure(figsize=(15, 8))
    pd.plotting.parallel_coordinates(parallel_data, 'school', color=sns.color_palette("Set2"))
    
    plt.title('Mối quan hệ giữa Điểm số, Thời gian học, Sức khỏe và Mức tiêu thụ Rượu')
    plt.xlabel('Các yếu tố')
    plt.ylabel('Giá trị (đã được chuẩn hóa)')
    plt.legend(title='Trường học')
    plt.show()
```

### Diễn giải kết quả

Biểu đồ tọa độ song song cho thấy các mẫu hành vi và kết quả học tập của từng học sinh (mỗi đường là một học sinh) qua nhiều biến số khác nhau.

### Câu 8: Sự thay đổi điểm số trung bình qua các kỳ thi của các trường

**Mục tiêu:** Theo dõi sự tiến bộ hoặc thay đổi trong kết quả học tập của học sinh ở mỗi trường qua ba kỳ thi (G1, G2, G3). Biểu đồ đường giúp nhà quản lý dễ dàng so sánh xu hướng học tập giữa hai trường.

**Các bước thực hiện:**
1.  Sử dụng phương pháp `melt` để chuyển đổi cấu trúc dữ liệu từ dạng rộng (các cột G1, G2, G3 riêng biệt) sang dạng dài, giúp việc vẽ biểu đồ dễ dàng hơn.
2.  Tính điểm trung bình cho mỗi trường tại mỗi kỳ thi.
3.  Vẽ biểu đồ đường (`line plot`) để thể hiện sự thay đổi điểm số trung bình qua các kỳ thi, với mỗi đường đại diện cho một trường.

```python
if data is not None:
    # Sử dụng melt để chuyển đổi dữ liệu
    data_melted = data.melt(id_vars=["school"], value_vars=["G1", "G2", "G3"], 
                              var_name="Kỳ thi", value_name="Điểm số")
    
    # Vẽ biểu đồ đường
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=data_melted, x="Kỳ thi", y="Điểm số", hue="school", marker='o', errorbar='sd')
    
    plt.title('Sự thay đổi Điểm số Trung bình qua các Kỳ thi')
    plt.xlabel('Kỳ thi')
    plt.ylabel('Điểm số Trung bình')
    plt.ylim(0, 14) # Đặt giới hạn trục y để dễ so sánh hơn
    plt.legend(title='Trường học')
    plt.grid(True)
    plt.show()
```

### Câu 10: Ảnh hưởng của Tiêu thụ Rượu bia đến Điểm số & Tổng kết câu chuyện

**Mục tiêu:** Đánh giá mức độ ảnh hưởng của việc tiêu thụ rượu bia (cả trong tuần và cuối tuần) đối với điểm số cuối kỳ (G3) của học sinh. Phân tích này cung cấp một cảnh báo rõ ràng về tác động tiêu cực của thói quen này. Cuối cùng, tổng hợp các phân tích để kể một câu chuyện hoàn chỉnh cho nhà quản lý giáo dục.

**Các bước thực hiện:**
1.  Tạo ra một biến mới `total_alc` bằng cách cộng mức độ tiêu thụ rượu trong tuần (`Dalc`) và cuối tuần (`Walc`).
2.  Nhóm dữ liệu theo `total_alc` và tính điểm G3 trung bình cho mỗi nhóm.
3.  Sử dụng biểu đồ điểm (`point plot`) để thể hiện mối quan hệ này.

```python
if data is not None:
    # Tạo biến tổng mức tiêu thụ rượu
    data['total_alc'] = data['Dalc'] + data['Walc']
    
    # Tính điểm trung bình G3 theo tổng mức tiêu thụ rượu
    avg_g3_by_alc = data.groupby('total_alc')['G3'].mean().reset_index()
    
    # Vẽ biểu đồ
    plt.figure(figsize=(12, 7))
    sns.pointplot(data=avg_g3_by_alc, x='total_alc', y='G3', color='red')
    
    plt.title('Ảnh hưởng của Tổng mức Tiêu thụ Rượu đến Điểm Trung bình G3')
    plt.xlabel('Tổng mức tiêu thụ rượu (Trong tuần + Cuối tuần)')
    plt.ylabel('Điểm Trung bình G3')
    plt.grid(True)
    plt.show()
```

3.  **Cảnh báo về lối sống: Sức khỏe, chuyên cần và tiêu thụ rượu bia:**
    - **Câu chuyện:** Dữ liệu đã vẽ nên một bức tranh rõ ràng: sức khỏe tốt, đi học đều đặn và đặc biệt là tránh xa rượu bia có mối tương quan mạnh mẽ với điểm số cao. Việc tiêu thụ rượu bia là yếu tố tiêu cực rõ rệt nhất, kéo điểm số trung bình xuống một cách đáng báo động.
    - **Hành động đề xuất:** Đặt sức khỏe thể chất và tinh thần của học sinh làm ưu tiên. Theo dõi chặt chẽ tình hình chuyên cần để phát hiện các trường hợp cá biệt. Quan trọng nhất, cần có chương trình can thiệp và giáo dục quyết liệt về tác hại của rượu bia đối với học sinh.

> **Tóm lại, những phân tích này cung cấp cơ sở dữ liệu vững chắc để nhà quản lý giáo dục và nhà tâm lý học đường có thể triển khai các chương trình hỗ trợ và can thiệp hiệu quả hơn, không chỉ tập trung vào kiến thức trên lớp mà còn vào việc xây dựng thói quen học tập, lối sống lành mạnh và tăng cường sự kết nối với gia đình, từ đó giúp học sinh phát triển toàn diện và tối ưu hóa kết quả học tập của mình.**
